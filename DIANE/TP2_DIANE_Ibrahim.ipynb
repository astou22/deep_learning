# A) RÃ©gression avec un Perceptron Multi-Couches

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import *
from sklearn.decomposition import PCA
import sklearn.preprocessing as sklp
from sklearn.neural_network import MLPRegressor
from sklearn.neural_network import MLPClassifier
```

```python
x = np.linspace(0,8,20).reshape(-1,1)
y = np.sin(x) + np.random.randn(20)*0.2
```

```python
mlpreg = MLPRegressor(hidden_layer_sizes=3, activation='logistic', 
                   max_iter=100, learning_rate_init=0.1)
```

```python
mlpreg = mlpreg.fit(x, y[:,1])
estimation_y = mlpreg.predict(x)
```

```python
fig = plt.figure(figsize=(12, 8))
axes = plt.axes()
axes.scatter(x, y[:,1], label='fonction rÃ©elle', color='green', s=25)
axes.plot(x, y[:,1])
axes.scatter(x, estimation_y, label='fonction estimÃ©e', color='black', s=25)
axes.plot(x, estimation_y)
plt.xlabel("Axe x")
plt.ylabel("Axe y")
plt.title("Fonction sinusoide")
plt.legend()
plt.show()
```

# B) Classification avec un perceptron multi-couches

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
y = (iris.target != 0)*1
```

```python
mlpclass = MLPClassifier(hidden_layer_sizes=2, activation='logistic', 
                     learning_rate_init=0.2, solver='lbfgs')
mlpclass = mlpclass.fit(X, y)
estimation_y = mlpclass.predict(X)
```

```python
pcaX = PCA(2).fit_transform(X)
fig = plt.figure(figsize=(12, 8))
axes = plt.axes()
axe1 = fig.add_subplot(211)
axe2 = fig.add_subplot(212)
axe1.scatter(pcaX[:,0], pcaX[:, 1], c=y, s=25)
axe2.scatter(pcaX[:,0], pcaX[:, 1], c=estimation_y, s=25)
plt.title("MLPClassifier Iris")
plt.show()
```

```python
print(estimation_y)
print(y)
"""On remarque que l'estimateur est bon."""
```

```python
stand = sklp.StandardScaler()
X_normalized = stand.fit_transform(X)
```

```python
mlpclass = mlpclass.fit(X_normalized, y)
estimation_y = mlpclass.predict(X_normalized)
pcaX = PCA(2).fit_transform(X_normalized)
fig = plt.figure(figsize=(12, 8))
axes = plt.axes()
axe1 = fig.add_subplot(211)
axe2 = fig.add_subplot(212)
axe1.scatter(pcaX[:,0], pcaX[:,1], c=y, s=25)
axe2.scatter(pcaX[:,0], pcaX[:,1], c=estimation_y, s=25)
plt.title("MLPClassifier Iris NormalisÃ©")
plt.show()
```

```python
# Variation des paramÃ¨tres du MLP
mlpclass_2 = MLPClassifier(hidden_layer_sizes=9, activation='logistic', 
                     learning_rate_init=0.3, solver='lbfgs')
mlpclass_2 = mlpclass_2.fit(X_normalized, y)
estimation_y = mlpclass_2.predict(X_normalized)
pcaX = PCA(2).fit_transform(X_normalized)
fig = plt.figure(figsize=(12, 8))
axes = plt.axes()
axe1 = fig.add_subplot(211)
axe2 = fig.add_subplot(212)
axe1.scatter(pcaX[:,0], pcaX[:,1], c=y, s=25)
axe2.scatter(pcaX[:,0], pcaX[:,1], c=estimation_y, s=25)
plt.title("MLPClassifier_2 Iris NormalisÃ©")
plt.show()
```

```python
mlpclass_3 = MLPClassifier(hidden_layer_sizes=15, activation='logistic', 
                     learning_rate_init=0.01, solver='lbfgs')
mlpclass_3 = mlpclass_3.fit(X_normalized, y)
estimation_y = mlpclass_3.predict(X_normalized)
pcaX = PCA(2).fit_transform(X_normalized)
fig = plt.figure(figsize=(12, 8))
axes = plt.axes()
axe1 = fig.add_subplot(211)
axe2 = fig.add_subplot(212)
axe1.scatter(pcaX[:,0], pcaX[:,1], c=y, s=25)
axe2.scatter(pcaX[:,0], pcaX[:,1], c=estimation_y, s=25)
plt.title("MLPClassifier_3 Iris NormalisÃ©")
plt.show()
```

# C) RÃ©duction de dimensions avec un Perceptron Multi-couches

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
mlp_train = MLPClassifier(hidden_layer_sizes=2, activation='logistic', 
                     learning_rate_init=0.2)
pcaX = PCA(2).fit_transform(X)
mlp_train = mlp_train.fit(pcaX, y)
estimation_y = mlp_train.predict(pcaX)
print(estimation_y)
print(y)

print(mlp_train.coefs_)
```
